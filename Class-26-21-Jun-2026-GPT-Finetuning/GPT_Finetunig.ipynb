```python
clinet.fine_tuning.jobs.create(
  training_file="file-2G95XDfh8PN4nH9M1KnJtg",
  model="gpt-4o-2024-08-06",
  suffix="genaibootcamp-live-class",
  method={
    "type": "supervised",
    "supervised": {
      "hyperparameters": {
        "learning_rate_multiplier": 1.0,
        "n_epochs": 3
      }
    }
  }
)
```

```python
clinet.chat.completions.create(
    model="ft:gpt-4o-2024-08-06:personal:genaibootcamp-live-class:DtEzD4Nc",
    messages=[
        {"role": "system", "content": "You are a customer support agent for a smartphone company whose primary goal is to help users with issues they are experiencing with their smartphones. You are friendly and concise. You only provide factual answers to queries, and do not provide answers that are not related to smartphones."},
        {
            "role": "user",
            "content": "What warranty does a smartphone come with?"
        }
    ]
)
```

```python
answer = clinet.chat.completions.create(
    model="ft:gpt-4o-2024-08-06:personal:genaibootcamp-live-class:DtEzD4Nc",
    messages=[
        {"role": "system", "content": "You are a customer support agent for a smartphone company whose primary goal is to help users with issues they are experiencing with their smartphones. You are friendly and concise. You only provide factual answers to queries, and do not provide answers that are not related to smartphones."},
        {
            "role": "user",
            "content": "Can I use the same apps on Android and iPhone?"
        }
    ]
)
```

```python
print(answer.choices[0].message.content)
```

## this is my ground truth

"Not all apps are available on both platforms. Some apps are exclusive to either Android or iOS, depending on the app store."

## this is a model generated ans after finetuning

Apps need to be downloaded separately from Google Play Store for Android and the App Store for iPhone. Compatibility depends on the developer.

## response from the general purpose model

Many apps are available on both Android and iPhone. However, some apps might be exclusive to one platform or have different versions. It's best to check the app's availability in the Google Play Store and Apple App Store.
